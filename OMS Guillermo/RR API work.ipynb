```python
import os,sys,json
def check_local_dir(path = "."):
    os.makedirs(path,exist_ok=True)
    
def make_file(var,path=".",extension=".txt"):
    if extension != ".json":
        
        f = open(path+extension,"w")
        f.write(str(var))
        f.close()
    elif extension ==".json":
        json.dump(var,open(path+extension,"w"))
```
